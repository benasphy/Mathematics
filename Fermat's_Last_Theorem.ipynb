```python
# 1. checking Fermat's Last Theorem

def check_fermat(a, b, c, n):
    if n > 2 and a**n + b**n == c**n:
        print("Holy smokes, Fermat was wrong!")
    else:
        print("Nope, Fermat was right.")

# Example:
check_fermat(3, 4, 5, 3)  # Output: "Nope, Fermat was right."


```

```python
# 2. how fast a**n + b **n grows compared to c**n

import matplotlib.pyplot as plt
import numpy as np

def visualize_fermat(a, b, n_max, c_max):
    n_values = np.arange(3, n_max + 1)  # n > 2
    c_values = np.arange(1, c_max + 1)

    plt.figure(figsize=(12, 6))

    for n in n_values:
        a_n = a**n
        b_n = b**n
        lhs = a_n + b_n  # Left-hand side

        rhs = c_values**n  # Right-hand side for different c

        plt.plot(c_values, rhs, label=f'c^{n}')
        plt.axhline(y=lhs, color='r', linestyle='--', label=f'a^{n} + b^{n}' if n == 3 else None)

    plt.xlabel('c')
    plt.ylabel('Value')
    plt.title(f'Visualizing Fermat\'s Last Theorem for a={a}, b={b}')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example:
visualize_fermat(a=3, b=4, n_max=6, c_max=20)
```
